```python
"""date → تاريخ المباراة

home_team → الفريق صاحب الأرض

away_team → الفريق الضيف

team → الفريق الذي سجل الهدف

scorer → اسم اللاعب الذي سجل الهدف

minute → الدقيقة التي سجل فيها الهدف

own_goal → هل الهدف هدف عكسي

TRUE = هدف في مرمى فريقه

FALSE = هدف طبيعي

penalty → هل الهدف من ضربة جزاء

TRUE = ركلة جزاء

FALSE = هدف من اللعب

"""
```
